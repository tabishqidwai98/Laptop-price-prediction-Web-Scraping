```python
url ='https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
sc = Scraper(url)
```

```python
result = sc.get_repeating_page_data(
    target=Tag('body'),
    items=Tag(cls='tUxRFH'),
    title = Tag(cls='KzDlHZ'),
    price = Tag(cls='Nx9bqj _4b5DiR'),
    oprice = Tag(cls='yRaY8j ZYYwLA'),
    rating_n_review=Tag('span', cls='Wphh3N'),
    link = Tag('a', cls='CGtC98', output='href')
)
```

```python
import pandas as pd
df = pd.DataFrame(result)
df.to_csv('lappy1.csv', index=True)
```
